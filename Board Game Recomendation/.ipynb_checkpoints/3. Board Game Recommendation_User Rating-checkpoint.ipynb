## 3. Board Game Recommendation_User Rating
## Overview

The first model was built by user the user rating of the game and make a recommendation system based on the pairwise distance scores. There are limitations on data collected as there are no user ratings from a single user on multiple games, the recommendation system based on collabrative filtering is not applicable. Thus, the first recommendation system is purely based on the average rating of a board game to recommend the top 5 board games that have similar user ratings.

```python
import pandas as pd
import numpy as np

from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances,linear_kernel
```

## Cosine Similarity_User Rating

```python
# Read in the cleaned dataset
bg_rating = pd.read_csv(file + 'boardgames_ratings.csv')
bg_rating.shape
# 3M rows
```

```python
# Resample (no of rating >= 100) & (comment length >= 10)
bg_rating_cs = bg_rating[(bg_rating['no of rating']>=100) & (bg_rating['comment_len']>=10)]
bg_rating_cs.shape
```

```python
bg_rating_cs['name'].nunique()
```

```python
# Making pivot table
pivot = pd.pivot_table(bg_rating_cs, index='name', columns='user', values='rating')
pivot.head()
```

```python
pivot.shape
```

```python
sparse_pivot = sparse.csr_matrix(pivot.fillna(0))
print(sparse_pivot)
```

```python
# Creating a dataframe with the cosine similarity scores among games
recommender = pairwise_distances(sparse_pivot, metric='cosine')
recommender_df = pd.DataFrame(recommender, columns=pivot.index, index=pivot.index)
recommender_df.head()
```

```python
# Recommendation function based on the rating

def boardgames_recommender(search):
    print(search)

    # Average rating of the game
    print('Average rating : ', round(pivot.loc[search, :].mean(),2))
    # Total number of rating of the boardgame
    print('Number of ratings : ', pivot.T[search].count())
    print('')
    # 5 closest boardgames based on the distance score, lower the better
    print('5 closest boardgames : ')
    print(recommender_df[search].sort_values()[1:6])
    print('')
```

```python
boardgames_recommender("Pandemic")
```

The result of this recommendation system is to find the game that have similar average rating. It will show the average rating of the game and also number of ratings.
